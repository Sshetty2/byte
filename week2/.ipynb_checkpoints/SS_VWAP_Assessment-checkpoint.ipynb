```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('TSLA.csv')
```

```python
df
```

```python
df['Typical Price'] = (df.High + df.Low + df.Close)/3
```

```python
df
```

```python
df['Volume * Price'] = (df.Volume * df['Typical Price'])
```

```python
df
```

```python
df.cumsum()
```

```python
df['Running Total of Volume'] = df.Volume.cumsum()
```

```python
df
```

```python
df['Running Total of Volume * Price'] = df['Volume * Price'].cumsum()
```

```python
df
```

```python
df['VWAP'] = df['Running Total of Volume * Price']/df['Running Total of Volume']
```
